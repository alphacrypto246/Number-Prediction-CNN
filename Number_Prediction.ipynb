```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# It is a 28x28 pxl image
X_train[0].shape
```

```python
X_train
```

```python
plt.matshow(X_train[0])
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
# We will flatten our array to a single dimensional array

X_train_flatten = X_train.reshape(len(X_train), 28*28)
X_test_flatten = X_test.reshape(len(X_test), 28*28)
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(
    optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']
)

model.fit(X_train_flatten, y_train, epochs=10)
```

```python
# Evaluation

model.evaluate(X_test_flatten, y_test)
```

```python
# Checking out first number
plt.matshow(X_test[0])
```

```python
# Predicting the number through the model
y_predict = model.predict(X_test_flatten)

# Predicting our first number i.e. 7
y_predict[0]
```

```python
# This will give us our predicted number, which turns out to be 7
np.argmax(y_predict[0])
```

```python
y_test[:5]
```

```python
# Lets check our first 5 predictions from y_test

y_predict_label = [np.argmax(i) for i in y_predict]
y_predict_label[:5]
```

```python
# Confusion Matrix
cm = tf.math.confusion_matrix(y_test, np.argmax(y_predict, axis=1))
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# Adding a hidden layer

model_2 = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model_2.compile(
    optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']
)

model_2.fit(X_train_flatten, y_train, epochs=10)
```

```python
model_2.evaluate(X_test_flatten, y_test)
```

```python
y_predict_2 = model_2.predict(X_test_flatten)
```

```python
# Confusion Matrix
cm_2 = tf.math.confusion_matrix(y_test, np.argmax(y_predict_2, axis=1))
cm_2
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm_2, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# Adding a hidden layer

model_3 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(100, activation='tanh'),
    keras.layers.Dense(10, activation='sigmoid')
])

model_3.compile(
    optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']
)

model_3.fit(X_train, y_train, epochs=10)
```

```python
model_3.evaluate(X_test, y_test)
```

```python
y_predict_3 = model_3.predict(X_test)
```

```python
cm_3 = tf.math.confusion_matrix(y_test, np.argmax(y_predict_3, axis=1))
cm_3
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm_3, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
